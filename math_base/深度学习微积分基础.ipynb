### 积分和差乘除公式
$若\quad y(x) = f(x) + g(x), \quad 则 y^{'}(x) = \dfrac{dy}{dx} = \dfrac{df}{dx} + \dfrac{dg}{dx}$  
$若\quad y(x) = f(x) \cdot g(x), \quad 则 y^{'}(x) = \dfrac{dy}{dx} = f(x) \cdot \dfrac{dg}{dx} + g(x) \cdot \dfrac{df}{dx} \quad (若是除法则是相减)$  
$若复合函数\quad y(x) = f(g(x)), \quad 则 \quad y^{'} = f^{'}(g(x)) \cdot g^{'}(x) = \dfrac{df}{dg} \cdot \dfrac{dg}{dx}$

### 微分范例
$若\quad f(x) = x^{2}, \quad则\quad\frac{df}{dx} = \lim \limits_{\Delta x \to 0}\frac{f(x + \Delta x) - f(x)}{\Delta x} = \lim\frac{(x + \Delta x)^{2} - x^{2}}{\Delta x} = 2x + \lim \limits_{\Delta x \to 0}\Delta x = 2x$  
### 积分范例
$若\quad f(x) = x, \quad则\quad 积函数 F(x) = \dfrac{x^{2}}{2}, \quad\int_{0}^{1}f(x) dx = F(1) - F(0) = \dfrac{1^{2}}{2} - \dfrac{0^{2}}{2} = \dfrac{1}{2}$  

### 泰勒公式
设 n 是一个正整数. 如果定义在一个包含 a 的区间上的函数 f 在 a 点处 n+1 次可导，那么对于这个区间上的任意 x 都有：  
$f(x) = \sum\limits_{n=0}^{N}\dfrac{f^{n}(a)}{n!}(x - a) + R_{n}(x),\quad 其中的多项式称为函数在 a 处的泰勒展开式, R_ n(x) 是泰勒公式的余项且是 (x-a)^ n 的高阶无穷小, 可忽略. $
$当x在0时更简洁可得, \quad f(x) = \sum\limits_{n=0}{N}\dfrac{f^{(n)}}{n!}x^{n} = f(0) + f^{'}(0)x + \dfrac{f^{''}(0)}{2!}x^{2} + \cdots + \dfrac{f^{(n)}(0)}{n!}x^{n}$

### 牛顿法 ---- 二阶公式( 求最值 )
$f(x) = f(x_{0}) + (x - x_{0})f^{'}(x_{0}) + \frac{1}{2}(x - x_{0})^{2}f^{''}(x_{0})\quad 求能使f^{'}(x) = 0的解$  
$\because f^{'}(x) = f(x_{0}) + (x - x_{0})f^{''}(x_{0}) = 0$  
$\therefore x_{1} = x = x_{0} - \dfrac{ f^{'}(x_{0})}{f^{''}(x_{0})}$  
$\therefore 扩展至多维空间 \quad X_{n+1} = X_{n} - H(X_{n})^{-1}\Delta(X_{n})\quad H(X_{0})为Hessian矩阵$
